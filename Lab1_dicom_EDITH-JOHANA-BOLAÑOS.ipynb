### Create a dictionary `ewin` to summarize the information of the different energy windows. For this, complete the following `for` loop for the different keys of the window.

```python
# Inicializar un diccionario para las ventanas de energía
nwin = 1  # Definir el número de ventanas (en este caso, solo 1)
ewin = {}

for w in range(nwin):
    ewin[w] = {}  # Crear un diccionario vacío para cada ventana
    ewin[w]['lower'] = 435
    ewin[w]['upper'] = 650
    ewin[w]['center'] = (ewin[w]['lower'] + ewin[w]['upper']) / 2  # Calcular el centro
    ewin[w]['width'] = ewin[w]['upper'] - ewin[w]['lower']  # Calcular el ancho

# Mostrar el resultado
print(ewin)
```

## The sensitivity is defined as:

$S = \frac{C}{A\times t} $

where:

$S$ is the sensitivity

$C$ is the counts measured by the detector

$A$ is the activity of the source at the start of acquisition

$t$ is the acquisition duration
